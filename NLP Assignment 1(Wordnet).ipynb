```python
#load the wordnet
import nltk
nltk.download("wordnet")
from nltk.corpus import wordnet
```

```python
#words can have multiple senses so look at all senses for the word 'computer' & 'keyboard'
print("Finding synset for word computer" )
synsets = wordnet.synsets('computer')
print(synsets)
print("Finding synset for word keyboard" )
synsets = wordnet.synsets('keyboard')
print(synsets)
```

```python
#define words for which similarity need to find
com1 = wordnet.synset('computer.n.01')
com2 = wordnet.synset('calculator.n.01')
key1 = wordnet.synset('keyboard.n.01')
key2 = wordnet.synset('keyboard.n.02')
#print path similarity
print(wordnet.path_similarity(com1,key1))
print(wordnet.path_similarity(com1,key2))
print(wordnet.path_similarity(com2,key1))  
print(wordnet.path_similarity(com2,key2))
```

```python
#print leacock-chodorow similarity
print(wordnet.lch_similarity(com1,key1))
print(wordnet.lch_similarity(com1,key2))
print(wordnet.lch_similarity(com2,key1))
print(wordnet.lch_similarity(com2,key2))
```

```python
#print wu-palmer similarity
print(wordnet.wup_similarity(com1,key1))
print(wordnet.wup_similarity(com1,key2))
print(wordnet.wup_similarity(com2,key1))
print(wordnet.wup_similarity(com2,key2))
```

```python
list1 = ['computer','jerusalem','planet','canyon','opec','day','day','country','planet','environment','maradona','opec','money','computer','law','weather','network','nature','fbi','money','psychology','news','war','physics','bank','stock','planet','cerdit','hotel','closet','soap','planet','planet','movie','treatment','baby','money','television','psychology','game','admission','jerusalem','arafat','boxing','computer','money','tennis','telephone','currency','psychology']
list2 = ['keyboard','israel','galaxy','landscape','country','summer','dawn','citizen','people','ecology','football','oil','bank','software','lawyer','forecast','hardware','environment','investigation','wealth','freud','report','troops','proton','money','market','constellation','card','reservation','clothes','opera','astronomer','space','theater','recovery','mother','deposit','film','mind','team','ticket','palestinian','terror','round','internet','property','racket','communication','market','cognition']
for i in range (len(list1)):
    word1,word2=(list1[i],list2[i])
    synsets1=wordnet.synsets(word1,'n')
    synsets2=wordnet.synsets(word2,'n')
    print("1st list=",synsets1)
    print("2nd list=",synsets2)
    p=[]
    l=[]
    w=[]
    for x,y in [(x,y) for x in synsets1 for y in synsets2]:
        print(x,y)
        s1 = wordnet.path_similarity(x,y,simulate_root=False)
        p.append(s1)
    print("path_score=",p)
    if p:
        print("max_p =", max(p))
    else:
        print("max_p = None")
    #print("max p=",max(p) or 'None')
    for x,y in [(x,y) for x in synsets1 for y in synsets2]:
        s2 = wordnet.lch_similarity(x,y,simulate_root=False)
        l.append(s2)
    print("lch_score=",l)
    if l:
        print("max_l=",max(l))
    else:
        print("max_l= None")
    for x,y in [(x,y) for x in synsets1 for y in synsets2]:
        s3 = wordnet.wup_similarity(x,y,simulate_root=False)
        w.append(s3)
    print("wup_score=",w)
    if w:
        print("max w=",max(w))
    else:
        print("max_w = None")
    
```
